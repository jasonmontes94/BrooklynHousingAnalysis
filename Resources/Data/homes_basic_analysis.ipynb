```python
import matplotlib.pyplot as plt

# Load the data into a DataFrame

# Extract the sale_price column
sale_price = df['sale_price']

# Create a scatterplot of the sale_price column
plt.scatter(range(len(sale_price)), sale_price)
# Create a scatterplot with larger markers and a red color
plt.scatter(range(len(sale_price)), sale_price, marker='o', c='red', s=50)

# Set the x-axis label
plt.xlabel('Index')

# Set the y-axis label
plt.ylabel('Sale Price')

# Set the x-axis and y-axis limits
plt.xlim(0, len(sale_price))
plt.ylim(0, sale_price.max())

# Add a title to the plot
plt.title('Sale Prices')
# Show the plot
plt.show()
```
